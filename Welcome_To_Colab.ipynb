```
# NAME: journal.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
def write_entry(filename, entry):
    file = open(filename, "a")
    file.write(entry + "\n")
    file.close()


def read_journal(filename):
    try:
        file = open(filename, "r")
        contents = file.read()
        file.close()

        print("--- Your Journal ---")
        print(contents)

    except FileNotFoundError:
        print("No journal entries found yet.")


# PROGRAM LOGIC
filename = "journal.txt"

while True:
    choice = input("(W)rite, (R)ead, or (Q)uit? ").lower()

    if choice == "w":
        entry = input("Enter your entry: ")
        write_entry(filename, entry)
        print("Entry saved.\n")

    elif choice == "r":
        read_journal(filename)

    elif choice == "q":
        print("Goodbye!")
        break

    else:
        print("Invalid choice. Try again.\n")
```
