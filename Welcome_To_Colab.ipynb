```
# NAME: journal.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
def write_entry(filename, entry):
    file = open(filename, "a")
    file.write(entry + "\n")
    file.close()


# PROGRAM LOGIC
filename = "journal.txt"

while True:
    choice = input("(W)rite, (R)ead, or (Q)uit? ").lower()

    if choice == "w":
        entry = input("Enter your entry: ")
        write_entry(filename, entry)
        print("Entry saved.\n")

    elif choice == "r":
        print("--- Your Journal ---")
        print("(Read function not added yet in midpoint)")

    elif choice == "q":
        break
```
